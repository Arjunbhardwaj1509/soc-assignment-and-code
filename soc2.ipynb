```python
import torch.nn.functional as F
xenc=F.one_hot(xs,num_classes=27).float()
```

```python
#randomly initialise 27 neurons weights each neuron recieves 27 inputs
g=torch.Generator().manual_seed(2147483647)
W=torch.randn((27,27),generator=g,requires_grad=True)
```

```python
for i in range(400):
    #forward pass
    logits=xenc@W
    counts=xenc@W.exp()
    prob=counts/counts.sum(1,keepdims=True)# the last two lines are basically called as softmax
    loss=-prob[torch.arange(num),ys].log().mean() + 0.01*(W**2).mean()
    print(loss.item())
    #backward pass
    W.grad=None # set to zero to gradient
    loss.backward()
    #update
    W.data+=-30.0*W.grad


```
